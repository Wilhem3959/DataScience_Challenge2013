```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import collections
import csv
from itertools import groupby 
```

```python
df = pd.DataFrame(pd.read_csv('hubway_trips.csv'))
```

```python
df_1 = df['strt_statn']
df_2 = df['end_statn']
df_fusion = df_1.append(df_2)
```

```python
fusion = df_fusion.value_counts()
f_df = pd.DataFrame(fusion)

fusion_filter = df_fusion.value_counts() < 60000
graft = f_df[fusion_filter]
```

```python
graft.plot.bar(figsize = [25,10],legend = None)
```
