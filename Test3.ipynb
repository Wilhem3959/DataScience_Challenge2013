```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import collections
import csv
from itertools import groupby 
```

```python
df = pd.DataFrame(pd.read_csv('hubway_trips.csv'))

df['start_date'] = pd.to_datetime(df.start_date) 
df['end_date'] = pd.to_datetime(df.end_date)
```

```python
Week_days = df.end_date.dt.weekday_name
days = list(dict.fromkeys(Week_days))
print(days)
```

```python
counter = collections.Counter(Week_days)
count = counter.values()
print(counter)
print(count)
```

```python
plt.rcParams["figure.figsize"] = (20,16) 

fig, ax = plt.subplots() 

ax.bar(days,count);
```
