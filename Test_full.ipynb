```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import collections 
import csv
from itertools import groupby
```

```python
#Duration 
#a) Analysis of data: 
#given the data that we had, we were asked to look for the amount of people who were registered and who were casuals
#b) Approach used to generate the answer: 
#we extracted the data from the file given to us as dataframes and searched for a value count and graphed the results in order to find the answer we were searching for.
```

```python
#c) Python source code
df = pd.DataFrame(pd.read_csv('hubway_trips.csv'))
Subs = df['subsc_type'].value_counts()
register = df['subsc_type'] == 'Registered'
casual = df['subsc_type'] == 'Casual'

df_register = df[register]
df_casual = df[casual]

ave_register = df_register['duration']
ave_casual = df_casual['duration']

Subscription = ave_register.append(ave_casual).value_counts() >30000
subscription = pd.DataFrame(Subscription)
S = subscription[Subscription]

Subs.plot.bar(figsize = [18,10], color = "C9")
plt.title('Subscriptions', fontsize = 30)
plt.xlabel('Types of subscriptions', fontsize = 30)
plt.ylabel('Amount', fontsize = 30)
plt.xticks(fontsize = 15, rotation = 30)
plt.yticks(fontsize = 15)
plt.show()

Ave_register = df_register['duration'].mean()
Ave_casual = df_casual['duration'].mean()
print('Promedio de tiempo de usuarios registrados: '+str(Ave_register))
print('Promedio de tiempo de usuarios casuales: '+str(Ave_casual))

print('El numero de rentas de los usuarios registrados: '+str(df_register.shape[0]))
print('El numero de rentas de los usuarios casuales: '+str(df_casual.shape[0]))

#d) data visualization
#e) Answer/Conclusion: 
#The average trip duration for registered members is 733 and the average trip duration for casual members is 2293, by looking at the resultys we can conclude that the casual members had bigger trip durations than the registered members.
```

```python
#Start date
#a) Analysis of data: 
#given the data, we were asked to look for the peak hours of the Hubway Station
#b) Approach used to generate the answer: 
#we extracted the data of the hubways working hours from the file given to us as dataframes and searched for a value count of each of them and graphed the results in order to find the answer we were searching for.
```

```python
#c) python source code
df = pd.DataFrame(pd.read_csv('hubway_trips.csv'))

df['start_date'] = pd.to_datetime(df.start_date)
df['end_date'] = pd.to_datetime(df.end_date)
```

```python
hora = df.start_date.dt.hour
print(hora.mode())
hours = list(dict.fromkeys(hora))
print(hours)

counter = collections.Counter(hora)
count = counter.values()

plt.rcParams["figure.figsize"] = (24,16)
fig,ax = plt.subplots()
plt.title('Hora de uso', fontsize = 30)
plt.xlabel('horas', fontsize = 30)
plt.ylabel('Usuario', fontsize = 30)
ax.bar(hours,count);

#d) data visualization
#e) Answer/Conclusion: 
#The peak hours after graphing the data we could see that were between 15 and 20  hours.
```

```python
#End date
#a) Analysis of data: 
#After looking at the  data, we were asked to look which day of the week got the most hubway traffic.
#b) Approach used to generate the answer: 
#we extracted the data of the hubways weekly data and changed it as dataframes and searched for a value count of each of them and graphed the results in order to find the answer.
```

```python
#c python source code
Week_days = df.end_date.dt.weekday_name
days = list(dict.fromkeys(Week_days))
print(days)

counter1 = collections.Counter(Week_days) 
count1 = counter1.values()

plt.rcParams["figure.figsize"] = (24,16)
fig,ax = plt.subplots()
plt.title('Week Days', fontsize = 30)
plt.xlabel('Dias',fontsize = 30)
plt.ylabel('Usuarios',fontsize = 30)
ax.bar(days,count1);

#d) data visualization
#e) Answer/Conclusion: 
#After getting the visualization of the data we concluded that the days of the week who had the most hubway traffic were Wednesday and Thursday.
```

```python
#Start Station
#a) Analysis of data: 
#We where asked to search for the most popular stations and which stations made the most popular origin/destination pairs.
#b) Approach used to generate the answer: 
#After extracting the data and changed it to dataframes we graphed two different bar graphs with the results in order to find the answer.
```

```python
#End Station
#a) Analysis of data: 
#We where asked to search for the most Assymetric stations.
#b) Approach used to generate the answer: 
#After extracting the data and changed it to dataframes we graphed  the results in order to find the answer.
```

```python
#Gender
#a) Analysis of data: 
#We where asked to compare the different top stations for male and female members.
#b) Approach used to generate the answer: 
#After reading the data we graphed a bar graph with genders given to us of all the members to find the top stations.
```

```python
#c) python source code

Generos = pd.DataFrame(pd.read_csv('hubway_trips.csv'))

Men = Generos['gender'] == 'Male'
Women = Generos['gender'] == 'Female'

men = Generos[Men]
women = Generos[Women]

male_genders = men['gender']
female_genders = women['gender']

genders = male_genders.append(female_genders).value_counts()

Genders = pd.DataFrame(genders)

Genders_count = male_genders.append(female_genders).value_counts() > 30000
Genders2 = Genders[Genders_count]



Genders2.plot.bar(figsize = [25,10], legend = None)
plt.title('Generos',fontsize = 30)
plt.xlabel('Generos', fontsize = 30)
plt.ylabel('Cantidad',fontsize = 30)
plt.xticks(fontsize = 20, rotation = 30)
plt.yticks(fontsize = 20)
plt.show()

#d) data visualization
#e) Answer/Conclusion: 
#After graphing the data we found the difference of the male and female members in the stations.
```
